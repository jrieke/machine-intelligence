```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

"example in which a multilayer perceptron (MLP) can distinguish between
two classes, but a single connectionist neuron can not": XOR

```python
for i in range(50):
    a = np.random.normal(0, 2, 10)  # 10 elements, ie 1 for each hidden neuron
    w = np.random.normal(0, 1, 10)
    b = np.random.uniform(-2, 2, 10)
    
    x = np.linspace(-2, 2)
    y = np.sum(w * np.tanh(a * (x.reshape(-1, 1) - b)), axis=1)
    plt.plot(x, y)
    
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-10, 10)  # fix y range in all plots
```

```python
for i in range(50):
    a = np.random.normal(0, 0.5, 10)    
    w = np.random.normal(0, 1, 10)
    b = np.random.uniform(-2, 2, 10)
    
    x = np.linspace(-2, 2)
    y = np.sum(w * np.tanh(a * (x.reshape(-1, 1) - b)), axis=1)
    plt.plot(x, y)
    
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-10, 10)  # fix y range in all plots
```

The y values for b) have a smaller range. Also, the input-output curves look smoother. Both effects can be explained by the reduced values for a (the absolute values of the product inside tanh are smaller, tanh is monotically increasing, therefore the abolute value of y is smaller; a higher value of a introduces stronger "perturbations" of the input).

```python
import sys
best_mse = sys.maxint
best_y = None

x = np.linspace(-2, 2).reshape(-1, 1)
    
for i in range(50):
    a = np.random.normal(0, 2, 10)    
    w = np.random.normal(0, 1, 10)
    b = np.random.uniform(-2, 2, 10)
    
    y = np.sum(w * np.tanh(a * (x - b)), axis=1)
    mse = np.mean((y - (-x))**2)
    if mse < best_mse:
        best_mse = mse
        best_y = y
        
plt.plot(x, best_y, label='MLP')
plt.plot(x, -x, label='- x')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.ylim(-10, 10)  # fix y range in all plots
```

```python
import sys
min_mse = sys.maxint
best_y = None

x = np.linspace(-2, 2).reshape(-1, 1)
    
for i in range(50):
    a = np.random.normal(0, 0.5, 10)    
    w = np.random.normal(0, 1, 10)
    b = np.random.uniform(-2, 2, 10)
    
    y = np.sum(w * np.tanh(a * (x - b)), axis=1)
    mse = np.mean((y - (-x))**2)
    if mse < min_mse:
        min_mse = mse
        best_y = y
        
plt.plot(x, best_y, label='MLP')
plt.plot(x, -x, label='- x')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.ylim(-10, 10)  # fix y range in all plots
```
